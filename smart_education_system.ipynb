```python
# ========================================
# Smart Education System - Python Version
# ========================================
courses = [
    "Introduction to AI",
    "Data Science for Beginners",
    "Web Development Basics",
    "Cybersecurity Essentials",
    "Python Programming"
]

course_contents = [
    "AI simulates human intelligence. Used in robotics, automation, etc.",
    "Data Science analyzes data using Python, ML, statistics.",
    "Web Dev builds websites using HTML, CSS, JavaScript.",
    "Cybersecurity protects systems using encryption, firewalls.",
    "Python is a versatile programming language used in many fields."
]

username = "student"
password = "pass123"
enrolled = [False] * len(courses)
progress = [0] * len(courses)
latest_quiz_score = 0

# --- Login System ---
def login():
    print("=== Smart Education System Login ===")
    for _ in range(3):
        user = input("Username: ")
        pwd = input("Password: ")
        if user == username and pwd == password:
            print(f"Login successful! Welcome, {username}!")
            return True
        else:
            print("Invalid credentials. Try again.")
    print("Too many failed attempts. Exiting...")
    return False

# --- Courses ---
def show_courses():
    print("\n--- Available Courses ---")
    for i, course in enumerate(courses):
        print(f"{i+1}. {course}")

def enroll_course():
    show_courses()
    choice = int(input("Enter course number to enroll: ")) - 1
    if 0 <= choice < len(courses):
        if enrolled[choice]:
            print(f"You are already enrolled in '{courses[choice]}'")
        else:
            enrolled[choice] = True
            print(f"Successfully enrolled in '{courses[choice]}'")
    else:
        print("Invalid course number.")

def learn_course():
    show_courses()
    choice = int(input("Enter course number to learn: ")) - 1
    if 0 <= choice < len(courses):
        if enrolled[choice]:
            print(f"\n--- Learning: {courses[choice]} ---")
            print(course_contents[choice])
        else:
            print("Please enroll in the course first.")
    else:
        print("Invalid course number.")

# --- Quizzes ---
def course_quiz():
    show_courses()
    choice = int(input("Enter course number to attempt quiz: ")) - 1
    score = 0
    if 0 <= choice < len(courses) and enrolled[choice]:
        print(f"\n--- Quiz for {courses[choice]} ---")
        if choice == 0:  # AI
            score += input("Q1: AI stands for? (a) Artificial Intelligence (b) Auto Internet: ").lower() == 'a'
            score += input("Q2: AI is used in? (a) Cooking (b) Robots: ").lower() == 'b'
        elif choice == 1:  # DS
            score += input("Q1: Data Science uses? (a) Excel (b) Python: ").lower() == 'b'
            score += input("Q2: DS helps in? (a) Guessing (b) Decision Making: ").lower() == 'b'
        elif choice == 2:  # Web
            score += input("Q1: HTML is used for? (a) Design (b) Structure: ").lower() == 'b'
            score += input("Q2: JS is for? (a) Interactivity (b) Layouts: ").lower() == 'a'
        elif choice == 3:  # Cyber
            score += input("Q1: Cybersecurity protects? (a) Photos (b) Data: ").lower() == 'b'
            score += input("Q2: Good passwords are? (a) 123456 (b) Complex: ").lower() == 'b'
        elif choice == 4:  # Python
            score += input("Q1: Python is a? (a) Snake (b) Programming Language: ").lower() == 'b'
            score += input("Q2: Python is used in? (a) AI (b) Only web: ").lower() == 'a'
        print(f"Score: {score}/2")
    else:
        print("Invalid course or not enrolled.")

def general_quiz():
    global latest_quiz_score
    print("\n--- General Quiz ---")
    score = 0
    score += input("Q1: AI stands for? (a) Auto Intellect (b) Artificial Intelligence: ").lower() == 'b'
    score += input("Q2: Python is for? (a) Cooking (b) Programming: ").lower() == 'b'
    latest_quiz_score = score
    print(f"Your Score: {score}/2")

def final_exam():
    print("\n--- Final Exam ---")
    score = 0
    score += input("Q1: Cybersecurity protects? (a) Devices (b) Songs: ").lower() == 'a'
    score += input("Q2: HTML is used for? (a) Web layout (b) Cooking: ").lower() == 'a'
    score += input("Q3: AI is used in? (a) Cars (b) Chairs: ").lower() == 'a'
    score += input("Q4: Python is good for? (a) Sleep (b) Data Science: ").lower() == 'b'
    score += input("Q5: Data Science finds? (a) Patterns (b) Animals: ").lower() == 'a'
    print(f"Final Exam Score: {score}/5")
    if score >= 3:
        print("Status: PASSED 🎉")
    else:
        print("Status: FAILED. Study more!")

# --- Progress ---
def update_progress():
    show_courses()
    choice = int(input("Enter course number to update progress: ")) - 1
    if 0 <= choice < len(courses) and enrolled[choice]:
        new_progress = int(input("Enter new progress (0-100): "))
        if 0 <= new_progress <= 100:
            progress[choice] = new_progress
            print("Progress updated.")
        else:
            print("Invalid progress value.")
    else:
        print("Invalid or not enrolled in course.")

# --- Dashboard ---
def dashboard():
    print("\n--- Student Dashboard ---")
    print(f"Username: {username}")
    for i, enrolled_course in enumerate(enrolled):
        if enrolled_course:
            print(f"{courses[i]} - Progress: {progress[i]}%")
    print(f"Last Quiz Score: {latest_quiz_score}/2")

# --- Main Menu ---
def main_menu():
    while True:
        print("\n--- Main Menu ---")
        print("1. View Courses")
        print("2. Enroll in Course")
        print("3. Learn Course Content")
        print("4. Take Course Quiz")
        print("5. General Quiz")
        print("6. Final Exam")
        print("7. Update Progress")
        print("8. View Dashboard")
        print("9. Exit")

        choice = input("Enter choice: ")

        if choice == '1':
            show_courses()
        elif choice == '2':
            enroll_course()
        elif choice == '3':
            learn_course()
        elif choice == '4':
            course_quiz()
        elif choice == '5':
            general_quiz()
        elif choice == '6':
            final_exam()
        elif choice == '7':
            update_progress()
        elif choice == '8':
            dashboard()
        elif choice == '9':
            print("Thank you for using Smart Education System!")
            break
        else:
            print("Invalid choice. Try again.")

# --- Run Program ---
if login():
    main_menu()
```
